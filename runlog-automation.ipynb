```python
from operator import itemgetter
from datetime import datetime

def sort_by_date_desc(array):
    # Convert date strings to datetime objects for sorting
    for item in array:
        item['Date'] = datetime.strptime(item['Date'], '%Y-%m-%d')

    # Sort array by date in descending order
    array.sort(key=itemgetter('Date'), reverse=True)

    # Convert datetime objects back to strings
    for item in array:
        item['Date'] = item['Date'].strftime('%Y-%m-%d')

    return array
```

```python
def sort_by_date_desc(df):
    # Convert date strings to datetime objects for sorting
    df['Date'] = pd.to_datetime(df['Date'])

    # Sort dataframe by date in descending order
    df_sorted = df.sort_values('Date', ascending=False)

    # Convert datetime objects back to strings
    df_sorted['Date'] = df_sorted['Date'].dt.strftime('%Y-%m-%d')

    return df_sorted
```

```python
df_sorted = sort_by_date_desc(df)
```

```python
def sort_dataframe_by_date_descending(df):
    # Convert 'date' column to datetime format if it's not already
    df['date'] = pd.to_datetime(df['date'])
    
    # Sort the DataFrame by the 'date' column in descending order
    sorted_df = df.sort_values(by='date', ascending=False)
    
    return sorted_df
```

```python
df_sorted = sort_dataframe_by_date_descending(df);
```

```python
from datetime import datetime, timedelta
from typing import Optional

def create_new_run_entry(last_entry: dict,
                         date: Optional[datetime] = None,
                         duration: Optional[timedelta] = None,
                         length: Optional[float] = None,
                         run_type: Optional[str] = None,
                         surface: Optional[str] = None,
                         sleepHours: Optional[float] = None,
                         sleepToBedTime: Optional[datetime] = None,
                         sleepWakeTime: Optional[datetime] = None,
                         runListenedTo: Optional[str] = None,
                         temperature: Optional[float] = None,
                         shoeAge: Optional[float] = None,
                         startTime: Optional[datetime] = None) -> dict:
    new_entry = last_entry.copy()
    new_entry['RunId'] = last_entry['RunId'] + 1  # Increment runId
    # Update fields with new values if provided, else default to None
    new_entry['Date'] = date
    new_entry['Duration'] = duration
    new_entry['Length'] = length
    new_entry['Type'] = run_type
    new_entry['Surface'] = surface
    new_entry['SleepHours'] = sleepHours
    new_entry['SleepToBedTime'] = sleepToBedTime
    new_entry['SleepWakeTime'] = sleepWakeTime
    new_entry['RunListenedTo'] = runListenedTo
    new_entry['Temperature'] = temperature
    new_entry['ShoeAge'] = shoeAge
    new_entry['StartTime'] = startTime
    return new_entry

# Example usage:
last_entry = {
    'RunId': 1492,
    # Other fields...
}

# Create a new entry (you can add more parameters as needed)
new_entry = create_new_run_entry(last_entry)
print(new_entry)
```
