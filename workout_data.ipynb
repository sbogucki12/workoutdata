```python


df = workout_data

barbell_bench = df['Barbell Bench']
print(barbell_bench)


tricep_extension = df['Seated Tricep Extension']
tricep_extension
    
```

```python

def rateOfChange(items):
    list = []
    for index in range(len(items)):    
        if index < len(items)-1:
            rate = items[index + 1] - items[index]
            list.append(rate)
    return list
            
bb_rate = rateOfChange(barbell_bench)
bb_rate
tri_rate = rateOfChange(tricep_extension)
print(bb_rate)
print(tri_rate)
```
